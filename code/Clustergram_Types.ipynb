```python
import pyarrow.parquet as pq
import pandas as pd
from clustergram import Clustergram
import sklearn
import cudf
import matplotlib.pyplot as plt
import pickle

```

# Import the data and Groups

```python
df = cudf.read_parquet('../data/OA_Input_london.parquet')
df = df.drop(columns=['utla22cd', 'utla22nm'])
# Import groups
gp = cudf.read_parquet('../data/LOAC_Group.parquet')
```

```python
df
```

```python
gp
```

```python
# Set index
df = df.set_index('OA')
gp = gp.set_index('OA')
```

```python
# Append clusters
df_merged = df.merge(gp, on=['OA'], how='left')
```

```python
df_merged
```

```python
# Count of Block Gropups by Cluster
df_merged.SG.value_counts()
```

```python
#Split DF and remove cluster
groups = df_merged.groupby(df_merged.SG)
df_merged_1 = groups.get_group("A")
df_merged_2 = groups.get_group("B")
df_merged_3 = groups.get_group("C")
df_merged_4 = groups.get_group("D")
df_merged_5 = groups.get_group("E")
df_merged_6 = groups.get_group("F")
df_merged_7 = groups.get_group("G")

#remove cluster column
del df_merged_1['SG']
del df_merged_2['SG']
del df_merged_3['SG']
del df_merged_4['SG']
del df_merged_5['SG']
del df_merged_6['SG']
del df_merged_7['SG']

#convert to pandas
df_merged_1= df_merged_1.to_pandas()
df_merged_2= df_merged_2.to_pandas()
df_merged_3= df_merged_3.to_pandas()
df_merged_4= df_merged_4.to_pandas()
df_merged_5= df_merged_5.to_pandas()
df_merged_6= df_merged_6.to_pandas()
df_merged_7= df_merged_7.to_pandas()
```

```python
cgram = Clustergram(range(1, 7), backend='sklearn',n_init=1000)
cgram.fit(df_merged_1)
cgram.plot(figsize=(12, 8))
plt.savefig('../clustergram_A.pdf')
```
